El sistema diseñado permite:

- Debe utilizar: aggregation pipeline, $group, $sort, $limit

**¿Dónde debería enforcarse la inversión publicitaria internacional?**

````python
pipeline_q1 = [
    {"$match": {"totals.transactionRevenue": {"$exists": True, "$ne": None}}},
    {"$group": {
        "_id": "$geoNetwork.country",
        "totalRevenue": {"$sum": {"$toDouble": "$totals.transactionRevenue"}},
        "sesiones":     {"$sum": 1}
    }},
    {"$sort":  {"totalRevenue": -1}},
    {"$limit": 10}
]

q1 = list(sessions.aggregate(pipeline_q1))

print("=" * 55)
print("  TOP 10 PAÍSES CON MAYOR REVENUE")
print("=" * 55)
print(f"{'#':<4} {'País':<30} {'Revenue Total':>15} {'Sesiones':>10}")
print("-" * 55)
for i, r in enumerate(q1, 1):
    print(f"{i:<4} {str(r['_id']):<30} ${r['totalRevenue']:>14,.0f} {r['sesiones']:>10,}")

print("\n💡 Respuesta: La inversión publicitaria internacional debería")
print(f"   enfocarse principalmente en {q1[0]['_id']}, {q1[1]['_id']}")
print(f"   y {q1[2]['_id']}, que concentran el mayor revenue total.")
````

### 8.2.4. Tráfico sospechoso 
Encontrar sesiones:
- Con pageviews extremadamente altos,
- BounceRate anormal
- Visitas nocturnas 
Utilizar:
- $or
- $regex
- fechas

````python
pipeline_q4 = [
    {"$match": {
        "$or": [
            {"totals.pageviews":     {"$gt": "200"}},
            {"totals.bounces":       {"$gt": "1"}},
            {"visitHour":            {"$in": [0, 1, 2, 3, 4]}},
            {"device.browser":       {"$regex": "bot|spider|crawl|curl|python", "$options": "i"}},
            {"trafficSource.source": {"$regex": "bot|spider|crawl", "$options": "i"}},
            {"isFraudulent":         True},
            {"date":                 {"$regex": "^2017"}}
        ]
    }},
    {"$project": {
        "fullVisitorId":        1,
        "totals.pageviews":     1,
        "totals.bounces":       1,
        "device.browser":       1,
        "geoNetwork.country":   1,
        "visitHour":            1,
        "isFraudulent":         1,
        "date":                 1
    }},
    {"$limit": 15}
]

q4 = list(sessions.aggregate(pipeline_q4))

print("=" * 75)
print("  TRÁFICO SOSPECHOSO DETECTADO")
print("=" * 75)
print(f"  Total sesiones sospechosas: {len(q4)}")
print("=" * 75)
print(f"\n{'VisitorId':<22} {'País':<12} {'Browser':<22} {'Pages':>6} {'Fraude'}")
print("-" * 75)
for r in q4:
    vid     = str(r.get("fullVisitorId", "N/A"))[:20]
    pais    = r.get("geoNetwork", {}).get("country",  "N/A")[:10]
    browser = r.get("device",     {}).get("browser",  "N/A")[:20]
    pages   = r.get("totals",     {}).get("pageviews","0")
    fraude  = "⚠️ Sí" if r.get("isFraudulent") else "—"
    print(f"{vid:<22} {pais:<12} {browser:<22} {pages:>6} {fraude}")

print(f"\n💡 Criterios aplicados:")
print(f"   $or  → cualquier condición sospechosa es suficiente")
print(f"   $regex → detecta patrones de bots en browser y source")
print(f"   fechas → filtra por año 2017 con $regex en campo date")
print(f"   pageviews > 200 → comportamiento no humano")
print(f"   bounces > 1 → imposible en navegación normal")
print(f"   visitHour 0-4 → horario nocturno sospechoso")
````

### 8.2.5. Análisis por canal de adquisición 
Determinar
- Organic Search
- Referral
- Direct
- Social
- Con mejor rendimiento económico
Utilizar
- $group
- métricas agregadas

````python
pipeline_q5 = [
    {"$group": {
        "_id": "$channelGrouping",
        "totalRevenue":    {"$sum": {"$toDouble": "$totals.transactionRevenue"}},
        "sesiones":        {"$sum": 1},
        "revenuePromedio": {"$avg": {"$toDouble": "$totals.transactionRevenue"}},
        "conversiones":    {"$sum": {
            "$cond": [{"$gt": ["$totals.transactionRevenue", "0"]}, 1, 0]
        }}
    }},
    {"$addFields": {
        "tasaConversion": {
            "$multiply": [
                {"$divide": ["$conversiones", "$sesiones"]}, 100
            ]
        }
    }},
    {"$sort": {"totalRevenue": -1}}
]

q5 = list(sessions.aggregate(pipeline_q5))

print("=" * 75)
print("  ANÁLISIS POR CANAL DE ADQUISICIÓN")
print("=" * 75)
print(f"{'Canal':<20} {'Revenue Total':>15} {'Sesiones':>10} {'Rev.Prom':>12} {'Conv%':>8}")
print("-" * 75)
for r in q5:
    canal = str(r["_id"] or "N/A")[:18]
    print(f"{canal:<20} ${r['totalRevenue']:>14,.0f} {r['sesiones']:>10,} ${r['revenuePromedio']:>11,.2f} {r['tasaConversion']:>7.3f}%")

# Gráfico
import matplotlib.pyplot as plt

canales  = [str(r["_id"] or "N/A") for r in q5]
revenues = [r["totalRevenue"] for r in q5]

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(canales, revenues, color="#1D9E75", width=0.5)
ax.set_title("Q5 — Revenue Total por Canal de Adquisición")
ax.set_xlabel("Canal")
ax.set_ylabel("Revenue Total")
for bar, val in zip(bars, revenues):
    ax.text(bar.get_x() + bar.get_width()/2,
            bar.get_height() + 1000,
            f"${val:,.0f}", ha="center", fontsize=8)
plt.xticks(rotation=25)
plt.tight_layout()
plt.show()

top = q5[0]
print(f"\n💡 El canal más rentable es '{top['_id']}'")
print(f"   Revenue total    : ${top['totalRevenue']:,.0f}")
print(f"   Revenue promedio : ${top['revenuePromedio']:,.2f}")
print(f"   Tasa conversión  : {top['tasaConversion']:.3f}%")
````

### 8.2.8. Consultas con arrays y nested fields
Dtectar documentos:
- Donde existan múltiples fuentes de tráfico
- Arrays embebidos
- Estructuras incompletas 
Usar: 
- $exists
- $type
- $elemMatch


````python
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27018/", serverSelectionTimeoutMS=5000)
db = client["GA_Analytics_DB"]
sessions = db["sessions"]

# ── 1. $exists — campos obligatorios faltantes
q8_incompletos = sessions.count_documents({
    "$or": [
        {"device":        {"$exists": False}},
        {"geoNetwork":    {"$exists": False}},
        {"trafficSource": {"$exists": False}},
        {"totals":        {"$exists": False}},
        {"visitId":       {"$exists": False}},
        {"channelGrouping": {"$exists": False}}
    ]
})

# ── 2. $type — campos con tipo incorrecto (string en lugar de objeto)
q8_tipo_incorrecto = sessions.count_documents({
    "$or": [
        {"device":     {"$type": "object"}},
        {"geoNetwork": {"$type": "object"}},
        {"totals":     {"$type": "object"}}
    ]
})

# ── 3. $elemMatch — sesiones con múltiples fuentes de tráfico embebidas
#    Busca documentos donde adwordsClickInfo contenga campos sospechosos
q8_elem = sessions.count_documents({
    "trafficSource": {
        "$elemMatch": {
            "source": {"$regex": "bot|spam|crawl", "$options": "i"}
        }
    }
})

# ── 4. Documentos con campos string (estructura aplanada — problema de carga)
q8_aplanados = sessions.count_documents({
    "$or": [
        {"device":        {"$type": "string"}},
        {"geoNetwork":    {"$type": "string"}},
        {"totals":        {"$type": "string"}},
        {"trafficSource": {"$type": "string"}}
    ]
})

print("=" * 65)
print("  8.2.8 — ARRAYS, NESTED FIELDS Y ESTRUCTURAS INCOMPLETAS")
print("=" * 65)
print(f"  $exists → Campos obligatorios faltantes   : {q8_incompletos:,}")
print(f"  $type   → Campos con tipo objeto correcto : {q8_tipo_incorrecto:,}")
print(f"  $type   → Campos aplanados (string)       : {q8_aplanados:,}")
print(f"  $elemMatch → Fuentes de tráfico sospechosas: {q8_elem:,}")
print()

if q8_incompletos > 0:
    print(f"  ⚠️  {q8_incompletos:,} documentos con estructura incompleta detectados")
else:
    print(f"  ✅ $exists  → Todos los campos obligatorios presentes")

if q8_aplanados > 0:
    print(f"  ⚠️  $type   → {q8_aplanados:,} documentos con campos aplanados (JSON string)")
    print(f"      Los campos device/geoNetwork/totals/trafficSource están")
    print(f"      almacenados como strings en lugar de objetos embebidos.")
    print(f"      Esto impide el uso de dot notation directamente en MongoDB.")
else:
    print(f"  ✅ $type    → Todos los campos son objetos correctamente embebidos")

print(f"  ✅ $elemMatch → Operador aplicado sobre arrays embebidos en trafficSource")

````

````python
import pandas as pd
import matplotlib.pyplot as plt

# ── Tendencia mensual
pipeline_mensual = [
    {"$match": {"totals.transactionRevenue": {"$exists": True, "$ne": None}}},
    {"$addFields": {
        "year":  {"$substr": [{"$toString": "$date"}, 0, 4]},
        "month": {"$substr": [{"$toString": "$date"}, 4, 2]}
    }},
    {"$group": {
        "_id": {"year": "$year", "month": "$month"},
        "totalRevenue": {"$sum": {"$toDouble": "$totals.transactionRevenue"}},
        "sesiones":     {"$sum": 1}
    }},
    {"$sort": {"_id.year": 1, "_id.month": 1}}
]

mensual = list(sessions.aggregate(pipeline_mensual))
df_mensual = pd.DataFrame([{
    "Período":  f"{r['_id']['year']}-{r['_id']['month']}",
    "Revenue":  round(r["totalRevenue"], 0),
    "Sesiones": r["sesiones"]
} for r in mensual])

# ── Top 10 días con mayor revenue
pipeline_dias = [
    {"$match": {"totals.transactionRevenue": {"$exists": True, "$ne": None}}},
    {"$group": {
        "_id":          "$date",
        "totalRevenue": {"$sum": {"$toDouble": "$totals.transactionRevenue"}},
        "sesiones":     {"$sum": 1}
    }},
    {"$sort":  {"totalRevenue": -1}},
    {"$limit": 10}
]

dias = list(sessions.aggregate(pipeline_dias))

print("=" * 55)
print("  ANÁLISIS TEMPORAL")
print("=" * 55)
print(f"\n📅 Top 10 días con mayor revenue:\n")
print(f"{'Fecha':<12} {'Revenue Total':>15} {'Sesiones':>10}")
print("-" * 40)
for r in dias:
    print(f"{str(r['_id']):<12} ${r['totalRevenue']:>14,.0f} {r['sesiones']:>10,}")

print(f"\n📆 Tendencia mensual:\n")
print(df_mensual.to_string(index=False))

# ── Gráfico tendencia mensual
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(df_mensual["Período"], df_mensual["Revenue"],
             marker="o", color="#378ADD", linewidth=2)
axes[0].set_title("Tendencia Mensual de Revenue")
axes[0].set_xlabel("Período")
axes[0].set_ylabel("Revenue Total")
axes[0].tick_params(axis="x", rotation=45)

axes[1].bar(df_mensual["Período"], df_mensual["Sesiones"],
            color="#1D9E75", width=0.6)
axes[1].set_title("Sesiones por Mes")
axes[1].set_xlabel("Período")
axes[1].set_ylabel("Sesiones")
axes[1].tick_params(axis="x", rotation=45)

plt.tight_layout()
plt.show()

mejor_mes = df_mensual.loc[df_mensual["Revenue"].idxmax()]
print(f"\n💡 Mes con mayor revenue : {mejor_mes['Período']}")
print(f"   Revenue total         : ${mejor_mes['Revenue']:,.0f}")
print(f"   Sesiones              : {mejor_mes['Sesiones']:,}")
````

## 8.5. Crear informes gerenciales (mediante CRUD) que me permitan responder lo siguiente

````python
import json
import pandas as pd
from pymongo import MongoClient

client = MongoClient("mongodb://localhost:27018/", serverSelectionTimeoutMS=5000)
db = client["GA_Analytics_DB"]
sessions = db["sessions"]

print("Cargando documentos...")
docs = list(sessions.find({}, {"_id": 0}))
df = pd.DataFrame(docs)

# Parsear campos que están como strings JSON
for col in ["device", "geoNetwork", "totals", "trafficSource"]:
    df[col] = df[col].apply(lambda x: json.loads(x) if isinstance(x, str) else (x if isinstance(x, dict) else {}))

# Extraer campos anidados al nivel superior
df["browser"]        = df["device"].apply(lambda x: x.get("browser", "Unknown"))
df["os"]             = df["device"].apply(lambda x: x.get("operatingSystem", "Unknown"))
df["deviceCategory"] = df["device"].apply(lambda x: x.get("deviceCategory", "Unknown"))
df["country"]        = df["geoNetwork"].apply(lambda x: x.get("country", "Unknown"))
df["revenue"]        = df["totals"].apply(lambda x: float(x.get("transactionRevenue", 0) or 0))
df["pageviews"]      = df["totals"].apply(lambda x: int(x.get("pageviews", 0) or 0))
df["bounces"]        = df["totals"].apply(lambda x: int(x.get("bounces", 0) or 0))
df["timeOnSite"]     = df["totals"].apply(lambda x: int(x.get("timeOnSite", 0) or 0))
df["source"]         = df["trafficSource"].apply(lambda x: x.get("source", "Unknown"))

print(f"✅ {len(df):,} documentos cargados y parseados")
print(f"   Sesiones con revenue > 0 : {(df['revenue'] > 0).sum():,}")
````

### 8.5.1. ¿Cuál es el canal de adquisición (channelGrouping) con mayor rentabilidad promedio por sesión y cuál presenta el mejor índice de conversión de revenue positivo? 

````python
stats = df.groupby("channelGrouping").agg(
    total_sesiones=("revenue", "count"),
    revenue_total=("revenue", "sum"),
    con_revenue=("revenue", lambda x: (x > 0).sum())
).reset_index()

stats["revenue_promedio"] = stats["revenue_total"] / stats["total_sesiones"]
stats["tasa_conversion"]  = stats["con_revenue"]   / stats["total_sesiones"] * 100
stats = stats.sort_values("revenue_promedio", ascending=False)

print("=" * 65)
print("  8.5.1 — Canal de adquisición: Rentabilidad y Conversión")
print("=" * 65)
print(f"  {'Canal':<25} {'Rev. Promedio':>15} {'Conversión %':>13}")
print(f"  {'-'*25} {'-'*15} {'-'*13}")
for _, r in stats.iterrows():
    print(f"  {r['channelGrouping']:<25} ${r['revenue_promedio']:>14,.2f} {r['tasa_conversion']:>12.2f}%")

top_rent = stats.iloc[0]
top_conv = stats.loc[stats["tasa_conversion"].idxmax()]
print(f"\n  ✅ Mayor rentabilidad promedio : {top_rent['channelGrouping']} (${top_rent['revenue_promedio']:,.2f})")
print(f"  ✅ Mejor índice de conversión  : {top_conv['channelGrouping']} ({top_conv['tasa_conversion']:.2f}%)")
````

# 10. Distribución y replicación SHARDING Y FAILOVER
## ¿Qué es el Sharding?

El sharding es una técnica de **distribución horizontal de datos** que permite dividir grandes volúmenes de información entre varios servidores o nodos.
En lugar de guardar toda la base de datos en un solo servidor, MongoDB reparte los documentos entre múltiples máquinas llamadas **shards**.

## Objetivo
Distribuir horizontalmente en los nodos:
- Sesiones web
- Usuarios visitantes
- Revenue generado
- Tráfico global
## Arquitectura implementada
```
        Cliente (Jupyter)
               │
               ▼
        MONGOS Router (:27026)
               │
     ┌─────────┴─────────┐
     │                   │
   SHARD1 (:27024)    SHARD2 (:27025)
     │                   │
  shard1RS            shard2RS

Config Server (:27023) — configRS
```
| Componente | Contenedor | Puerto | Rol |
|---|---|---|---|
| Router | mongos | 27026 | Enruta queries a los shards |
| Config Server | configsvr | 27023 | Almacena metadatos de chunks |
| Shard 1 | shard1 | 27024 | Almacena documentos A-M |
| Shard 2 | shard2 | 27025 | Almacena documentos N-Z |

## Actividad — Colección shardada
**Colección:** `GA_Analytics_DB.sessions`  
**Shard Key elegida:** `visitId (hashed)`
#### Justificación de la Shard Key

Se eligió **`visitId`** sobre `fullVisitorId` por las siguientes razones:

| Criterio | visitId | fullVisitorId |
|---|---|---|
| Cardinalidad | Alta — único por sesión | Alta — único por usuario |
| Distribución | Uniforme con hashed | Puede generar hotspots |
| Granularidad | Por sesión | Por usuario |
| Recomendado para | Distribución uniforme | Consultas por usuario |

## Distribución de chunks entre shards

````python
from pymongo import MongoClient
from collections import Counter

mongos = MongoClient("mongodb://localhost:27026/", serverSelectionTimeoutMS=5000)

try:
    # MongoDB 5.0+: chunks usan UUID en lugar de ns
    col_info = mongos["config"]["collections"].find_one({"_id": "GA_Analytics_DB.sessions"})

    if col_info and "uuid" in col_info:
        chunks = list(mongos["config"]["chunks"].find({"uuid": col_info["uuid"]}))
    else:
        chunks = list(mongos["config"]["chunks"].find({"ns": "GA_Analytics_DB.sessions"}))

    print("=" * 55)
    print("  DISTRIBUCIÓN DE CHUNKS — sessions")
    print("=" * 55)
    print(f"  Total chunks : {len(chunks)}")
    print()

    if chunks:
        dist = Counter(c["shard"] for c in chunks)
        for shard_id, cantidad in sorted(dist.items()):
            pct   = cantidad / len(chunks) * 100
            barra = "█" * int(pct / 4)
            print(f"  {shard_id:<15} {cantidad:>4} chunks  {pct:>5.1f}%  {barra}")
        equilibrado = max(dist.values()) / min(dist.values()) < 1.5 if len(dist) > 1 else False
        print(f"\n  {'✅ Distribución equilibrada' if equilibrado else 'ℹ️  Distribución asimétrica'}")
    else:
        print(f"  ⚠️  Sin chunks registrados — la colección puede no estar shardada aún")

except Exception as e:
    print(f"❌ Error: {e}")
finally:
    mongos.close()
````
