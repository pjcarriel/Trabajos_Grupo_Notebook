````python
from pymongo import MongoClient

# ── Cliente principal (CRUD normal)
client = MongoClient("mongodb://localhost:27018/", serverSelectionTimeoutMS=5000)
db = client["GA_Analytics_DB"]
sessions = db["sessions"]

# ── Cliente Replica Set (rs.status)
client_rs = MongoClient(
    "mongodb://localhost:27020/?directConnection=true",
    serverSelectionTimeoutMS=5000
)

# ── Cliente Mongos (sharding)
client_mongos = MongoClient(
    "mongodb://localhost:27026/?directConnection=true",
    serverSelectionTimeoutMS=5000
)

# ── Verificación rápida
print(f"✅ CRUD         → {client.server_info()['version']}")
print(f"✅ Replica Set  → {client_rs.admin.command('replSetGetStatus')['set']}")
print(f"✅ Sharding     → {len(client_mongos.admin.command('listShards')['shards'])} shards activos")
````

````python
import json
from pymongo import UpdateOne

def fix_json_fields():
    json_fields = ["device", "geoNetwork", "totals", "trafficSource"]
    bulk = []
    total = 0

    for doc in sessions.find({}, {"_id": 1, "device": 1, "geoNetwork": 1, "totals": 1, "trafficSource": 1}):
        update = {}
        for field in json_fields:
            val = doc.get(field)
            if isinstance(val, str):
                try:
                    update[field] = json.loads(val)
                except:
                    pass
        if update:
            bulk.append(UpdateOne({"_id": doc["_id"]}, {"$set": update}))

        if len(bulk) == 1000:
            sessions.bulk_write(bulk, ordered=False)
            total += len(bulk)
            bulk = []
            print(f"   Procesados: {total:,}", end="\r")

    if bulk:
        sessions.bulk_write(bulk, ordered=False)
        total += len(bulk)

    print(f"\n✅ Documentos corregidos: {total:,}")

fix_json_fields()
````
