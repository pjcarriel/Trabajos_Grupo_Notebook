El procesamiento eficiente de este volumen masivo de información requiere arquitecturas escalables, distribuidas y tolerantes a fallos capaces de mantener disponibilidad continua y soportar operaciones analíticas en tiempo real.

El presente proyecto tiene como finalidad implementar una solución basada en MongoDB distribuido utilizando técnicas de replicación, sharding y failover, aplicadas al dataset **Google Analytics Customer Revenue Prediction**. Además, se implementaron operaciones CRUD avanzadas, consultas gerenciales, detección de patrones fraudulentos y generación de indicadores estratégicos para la toma de decisiones empresariales.

El sistema diseñado permite:
- almacenar grandes volúmenes de datos,
- distribuir carga entre múltiples nodos,
- garantizar alta disponibilidad,

````python
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch
import matplotlib.patheffects as pe
import numpy as np

# ── Paleta de colores ──────────────────────────────────────
BG         = "#0F1117"
C_PURPLE   = "#7F77DD"   # cliente / jupyter
C_TEAL     = "#1D9E75"   # mongos
C_AMBER    = "#BA7517"   # config server
C_CORAL    = "#D85A30"   # shards
C_BLUE     = "#378ADD"   # primary
C_GRAY     = "#5F5E5A"   # secondary
C_GREEN    = "#639922"   # base de datos
WHITE      = "#F0EEE8"
MUTED      = "#888780"

# ── Construcción del grafo ─────────────────────────────────
G = nx.DiGraph()

nodes = {
    # (label,          color,    size,   shape)
    "Jupyter\nNotebook":        (C_PURPLE, 3200, "o"),
    "Mongos\nRouter":           (C_TEAL,   3200, "o"),
    "Config Server\nconfigRS":  (C_AMBER,  2800, "s"),
    "Shard 1\n27021":           (C_CORAL,  2600, "D"),
    "Shard 2\n27022":           (C_CORAL,  2600, "D"),
    "mongo1\nPRIMARY":          (C_BLUE,   3000, "o"),
    "mongo2\nSECONDARY":        (C_GRAY,   2600, "o"),
    "mongo3\nSECONDARY":        (C_GRAY,   2600, "o"),
    "GA_Analytics_DB\nsessions":(C_GREEN,  3200, "s"),
}

for n in nodes:
    G.add_node(n)

edges = [
    ("Jupyter\nNotebook",       "Mongos\nRouter",          "pymongo"),
    ("Mongos\nRouter",          "Config Server\nconfigRS", "metadatos"),
    ("Mongos\nRouter",          "Shard 1\n27021",          "chunks A-M"),
    ("Mongos\nRouter",          "Shard 2\n27022",          "chunks N-Z"),
    ("mongo1\nPRIMARY",         "mongo2\nSECONDARY",       "replicación"),
    ("mongo1\nPRIMARY",         "mongo3\nSECONDARY",       "replicación"),
    ("Shard 1\n27021",          "GA_Analytics_DB\nsessions","almacena"),
    ("Shard 2\n27022",          "GA_Analytics_DB\nsessions","almacena"),
    ("mongo1\nPRIMARY",         "GA_Analytics_DB\nsessions","escribe"),
]

for u, v, lbl in edges:
    G.add_edge(u, v, label=lbl)

# ── Posiciones manuales (layout en capas) ─────────────────
pos = {
    "Jupyter\nNotebook":         (4.0,  9.0),
    "Mongos\nRouter":            (4.0,  7.2),
    "Config Server\nconfigRS":   (4.0,  5.4),
    "Shard 1\n27021":            (1.8,  3.6),
    "Shard 2\n27022":            (6.2,  3.6),
    "mongo1\nPRIMARY":           (4.0,  3.6),
    "mongo2\nSECONDARY":         (2.2,  1.8),
    "mongo3\nSECONDARY":         (5.8,  1.8),
    "GA_Analytics_DB\nsessions": (4.0,  0.1),
}

# ── Figura ────────────────────────────────────────────────
fig, ax = plt.subplots(figsize=(13, 17), facecolor=BG)
ax.set_facecolor(BG)
ax.set_xlim(-0.2, 8.2)
ax.set_ylim(-0.8, 10.4)
ax.axis("off")

# ── Título ───────────────────────────────────────────────
ax.text(4.0, 10.15,
        "MongoDB — Arquitectura Distribuida",
        ha="center", va="center",
        fontsize=17, fontweight="bold", color=WHITE,
        fontfamily="monospace")
ax.text(4.0, 9.82,
        "GA_Analytics_DB  ·  Replica Set rs0  +  Sharding  +  Mongos",
        ha="center", va="center",
        fontsize=10, color=MUTED, fontfamily="monospace")

# ── Separadores de capa ──────────────────────────────────
layer_labels = [
    (9.45, "Capa 0 — Cliente"),
    (7.65, "Capa 1 — Router"),
    (5.85, "Capa 2 — Config"),
    (4.05, "Capa 3 — Shards  /  Primary"),
    (2.25, "Capa 4 — Secondaries"),
    (0.50, "Capa 5 — Persistencia"),
]
for y_sep, lbl in layer_labels:
    ax.axhline(y=y_sep - 0.52, color="#2A2D38", linewidth=0.8, linestyle="--", alpha=0.6)
    ax.text(0.0, y_sep - 0.42, lbl,
            fontsize=7.5, color="#4A4D58", va="bottom", fontfamily="monospace")

# ── Aristas ───────────────────────────────────────────────
edge_colors = {
    ("Jupyter\nNotebook",       "Mongos\nRouter"):           C_PURPLE,
    ("Mongos\nRouter",          "Config Server\nconfigRS"):  C_TEAL,
    ("Mongos\nRouter",          "Shard 1\n27021"):           C_CORAL,
    ("Mongos\nRouter",          "Shard 2\n27022"):           C_CORAL,
    ("mongo1\nPRIMARY",         "mongo2\nSECONDARY"):        C_BLUE,
    ("mongo1\nPRIMARY",         "mongo3\nSECONDARY"):        C_BLUE,
    ("Shard 1\n27021",          "GA_Analytics_DB\nsessions"):C_CORAL,
    ("Shard 2\n27022",          "GA_Analytics_DB\nsessions"):C_CORAL,
    ("mongo1\nPRIMARY",         "GA_Analytics_DB\nsessions"):C_BLUE,
}

for u, v, data in G.edges(data=True):
    x0, y0 = pos[u]
    x1, y1 = pos[v]
    color = edge_colors.get((u, v), MUTED)

    # Offset para evitar superposición en flechas paralelas
    dx = x1 - x0
    dy = y1 - y0
    norm = np.hypot(dx, dy) or 1
    px, py = -dy / norm * 0.08, dx / norm * 0.08

    # Curva suave
    ax.annotate(
        "", xy=(x1, y1), xytext=(x0, y0),
        arrowprops=dict(
            arrowstyle="-|>",
            color=color,
            lw=1.4,
            connectionstyle="arc3,rad=0.12",
            shrinkA=22, shrinkB=22,
        ),
        zorder=2
    )

    # Etiqueta de arista
    mx = (x0 + x1) / 2 + px * 1.4
    my = (y0 + y1) / 2 + py * 1.4
    ax.text(mx, my, data.get("label", ""),
            fontsize=7, color=color, ha="center", va="center",
            fontfamily="monospace", alpha=0.85,
            path_effects=[pe.withStroke(linewidth=2.5, foreground=BG)])

# ── Nodos ─────────────────────────────────────────────────
node_list   = list(nodes.keys())
color_list  = [nodes[n][0] for n in node_list]
size_list   = [nodes[n][1] for n in node_list]
marker_list = [nodes[n][2] for n in node_list]

for n in node_list:
    x, y = pos[n]
    color  = nodes[n][0]
    sz     = nodes[n][1]
    marker = nodes[n][2]
    r = np.sqrt(sz) / 290  # radio aprox en unidades del plot

    # Halo exterior (glow simulado con círculo grande translúcido)
    ax.scatter(x, y, s=sz * 2.2, c=color, alpha=0.10,
               marker=marker, zorder=3, linewidths=0)

    # Anillo exterior
    ax.scatter(x, y, s=sz * 1.15, c=BG, alpha=1.0,
               marker=marker, zorder=4, linewidths=0)

    # Nodo principal
    ax.scatter(x, y, s=sz, c=color, alpha=0.95,
               marker=marker, zorder=5,
               edgecolors=color, linewidths=1.5)

    # Etiqueta del nodo
    lines = n.split("\n")
    ax.text(x, y + 0.02,
            lines[0],
            ha="center", va="center",
            fontsize=9.5 if len(lines[0]) < 12 else 8.5,
            fontweight="bold", color=WHITE,
            fontfamily="monospace", zorder=6,
            path_effects=[pe.withStroke(linewidth=2, foreground=BG)])
    if len(lines) > 1:
        ax.text(x, y - 0.22,
                lines[1],
                ha="center", va="center",
                fontsize=7.5, color=color,
                fontfamily="monospace", zorder=6,
                path_effects=[pe.withStroke(linewidth=2, foreground=BG)])

# ── Leyenda ───────────────────────────────────────────────
legend_items = [
    (C_PURPLE, "Cliente / Jupyter"),
    (C_TEAL,   "Mongos router"),
    (C_AMBER,  "Config server"),
    (C_CORAL,  "Shards"),
    (C_BLUE,   "Nodo primary"),
    (C_GRAY,   "Nodo secondary"),
    (C_GREEN,  "Base de datos"),
]
handles = [
    mpatches.Patch(facecolor=c, edgecolor=WHITE, linewidth=0.5, label=lbl)
    for c, lbl in legend_items
]
legend = ax.legend(
    handles=handles,
    loc="lower left", bbox_to_anchor=(-0.01, -0.04),
    frameon=True, facecolor="#1A1D26", edgecolor="#2A2D38",
    fontsize=8, title="Componentes", title_fontsize=8.5,
    labelcolor=WHITE, ncol=2, columnspacing=1.2
)
legend.get_title().set_color(MUTED)

# ── Pie de figura ─────────────────────────────────────────
ax.text(4.0, -0.72,
        "shard key: visitId (hashed)  ·  replica set: rs0  ·  3 nodos  ·  2 shards  ·  1 mongos",
        ha="center", va="center",
        fontsize=7.5, color=MUTED, fontfamily="monospace")

plt.tight_layout()
plt.savefig("arquitectura_mongodb.png", dpi=180, bbox_inches="tight",
            facecolor=BG, edgecolor="none")
plt.show()
````
