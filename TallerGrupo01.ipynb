````python
import json
from pymongo import UpdateOne

def fix_json_fields():
    json_fields = ["device", "geoNetwork", "totals", "trafficSource"]
    bulk = []
    total = 0

    for doc in sessions.find({}, {"_id": 1, "device": 1, "geoNetwork": 1, "totals": 1, "trafficSource": 1}):
        update = {}
        for field in json_fields:
            val = doc.get(field)
            if isinstance(val, str):
                try:
                    update[field] = json.loads(val)
                except:
                    pass
        if update:
            bulk.append(UpdateOne({"_id": doc["_id"]}, {"$set": update}))

        if len(bulk) == 1000:
            sessions.bulk_write(bulk, ordered=False)
            total += len(bulk)
            bulk = []
            print(f"   Procesados: {total:,}", end="\r")

    if bulk:
        sessions.bulk_write(bulk, ordered=False)
        total += len(bulk)

    print(f"\n✅ Documentos corregidos: {total:,}")

fix_json_fields()
````

# 10. Distribución y replicación SHARDING Y FAILOVER
## ¿Qué es el Sharding?

El sharding es una técnica de **distribución horizontal de datos** que permite dividir grandes volúmenes de información entre varios servidores o nodos.
En lugar de guardar toda la base de datos en un solo servidor, MongoDB reparte los documentos entre múltiples máquinas llamadas **shards**.

## Objetivo
Distribuir horizontalmente en los nodos:
- Sesiones web
- Usuarios visitantes
- Revenue generado
- Tráfico global

## Actividad — Colección shardada
**Colección:** `GA_Analytics_DB.sessions`  
**Shard Key elegida:** `visitId (hashed)`
#### Justificación de la Shard Key

Se eligió **`visitId`** sobre `fullVisitorId` por las siguientes razones:

| Criterio | visitId | fullVisitorId |
|---|---|---|
| Cardinalidad | Alta — único por sesión | Alta — único por usuario |
| Distribución | Uniforme con hashed | Puede generar hotspots |
| Granularidad | Por sesión | Por usuario |
| Recomendado para | Distribución uniforme | Consultas por usuario |

---

## Pregunta crítica — ¿Por qué fullVisitorId podría generar hotspots?
Un **hotspot** ocurre cuando un shard recibe significativamente más carga
que los demás, desequilibrando el sistema.

`fullVisitorId` podría generar hotspots porque:

1. **Usuarios recurrentes** — un mismo `fullVisitorId` genera múltiples
   sesiones, concentrando documentos en el mismo chunk
2. **Distribución no uniforme** — si los IDs tienen un patrón secuencial,
   todos los documentos nuevos caen en el mismo shard
3. **Usuarios VIP** — los usuarios de alto valor generan más sesiones,
   sobrecargando el shard donde viven sus documentos

Por eso se eligió `visitId` con **hashed**, que garantiza distribución
aleatoria y uniforme entre shards sin importar el patrón del valor.

---

## Pregunta crítica — ¿Cómo afecta el sharding a UPDATE y DELETE?

| Operación | Sin shard key en filtro | Con shard key en filtro |
|---|---|---|
| UPDATE | Se ejecuta en TODOS los shards (broadcast) | Se ejecuta solo en el shard correcto |
| DELETE | Se ejecuta en TODOS los shards (broadcast) | Se ejecuta solo en el shard correcto |

**Impacto:**
- Sin shard key → operación lenta y costosa (scatter-gather)
- Con shard key → operación dirigida, rápida y eficiente (targeted)

**Recomendación:** siempre incluir `visitId` en los filtros de UPDATE
y DELETE para evitar operaciones broadcast.
